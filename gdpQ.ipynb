```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract csv into DataFrame
gdpQ_file = "resources/t10101q.csv"
gdpQ_df = pd.read_csv(gdpQ_file)
gdpQ_df.head()
```

```python
#Transpose DataFrame setting data as index
gdpQT_df =gdpQ_df.set_index('Factor').T
gdpQT_df.head()
```

```python
#Change index column name to date
gdpQT_df.columns.name="Date"
gdpQT_df.head()
```

```python
# View columns before creating filtered dataframe
gdpQT_df.columns
```

```python
# Create a filtered dataframe of GDP columns
gdpQ_col=["    Gross domestic product", "  Gross domestic product, current dollars"]
gdpQ_transformed_df = gdpQT_df[gdpQ_col]
gdpQ_transformed_df.reset_index(inplace=True)
gdpQ_transformed_df.head()
```

```python
#Change Date column name to ID
gdpQ_transformed_df.columns.name="ID"
gdpQ_transformed_df.head()
```

```python
# Change index column name to Date
gdpQ_transformed_df = gdpQ_transformed_df.rename(columns={'index':'Date','    Gross domestic product': 'Real GDP','  Gross domestic product, current dollars': 'Current GDP' })
gdpQ_transformed_df.head()
```

```python
#View columns of transformed dataframe
gdpQ_transformed_df.columns
```

```python
# Export file as a CSV, without the Pandas index, but with the header
gdpQ_transformed_df.to_csv("gdpQDB.csv", index=False, header=True)
```
